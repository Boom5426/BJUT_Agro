# 存为 Gene - 环境 - 性状 的成对数据

```python
import pandas as pd

ENV_path = 'normalized_ENV.xlsx'  # 很多 nan 值
Pheno_path = 'normalized_PHE.xlsx'

# Use pd.read_excel() for Excel files
ENV_CSV = pd.read_excel(ENV_path)
Pheno_CSV = pd.read_excel(Pheno_path)

print(ENV_CSV.shape, Pheno_CSV.shape)
```

```python
# 环境数据没有缺值
if ENV_CSV.isna().any().any():
    print("DataFrame 中存在 NaN 值")
else:
    print("DataFrame 中不存在 NaN 值")

# 检查每一列是否存在 NaN
nan_columns = ENV_CSV.columns[ENV_CSV.isna().any()].tolist()
print("包含 NaN 的列:", len(nan_columns))

# 检查每一行是否存在 NaN
nan_rows = ENV_CSV[ENV_CSV.isna().any(axis=1)]
print("包含 NaN 的行:\n", len(nan_rows))


# 检查整个 DataFrame 中 NaN 的总数
total_nans = ENV_CSV.isna().sum().sum()
print(f"整个 DataFrame 中 NaN 的总数: {total_nans}")
```

```python
# 检查表型数据是否存在 NaN 值
if Pheno_CSV.isna().any().any():
    print("DataFrame 中存在 NaN 值")
else:
    print("DataFrame 中不存在 NaN 值")

# 检查每一列是否存在 NaN
nan_columns = Pheno_CSV.columns[Pheno_CSV.isna().any()].tolist()
print("包含 NaN 的列:", len(nan_columns))

# 检查每一行是否存在 NaN
nan_rows = Pheno_CSV[Pheno_CSV.isna().any(axis=1)]
print("包含 NaN 的行:", len(nan_rows))

# 检查整个 DataFrame 中 NaN 的总数
total_nans = Pheno_CSV.isna().sum().sum()
print(f"整个 DataFrame 中 NaN 的总数: {total_nans}")

# 删除所有包含 NaN 的行
Pheno_CSV_cleaned = Pheno_CSV.dropna()

# 检查删除后的 DataFrame
print("\n删除包含 NaN 的行后的 DataFrame 形状:", Pheno_CSV_cleaned.shape)

# 验证删除后是否还有 NaN 值
if Pheno_CSV_cleaned.isna().any().any():
    print("删除后 DataFrame 中仍存在 NaN 值")
else:
    print("删除后 DataFrame 中不存在 NaN 值")

Pheno_CSV_cleaned
```

```python
# 加载数据
all_data = np.load('BJUT_all.npy', allow_pickle=True)

# 定义一个函数，用于替换非数值为 0
def replace_non_numeric_with_zero(data):
    # 创建一个与 data 形状相同的数组，初始化为 0
    processed_data = np.zeros_like(data, dtype=np.float32)
    
    # 遍历每个元素
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            # 检查元素是否为数值类型
            if isinstance(data[i, j], (int, float)):
                processed_data[i, j] = data[i, j]  # 保留数值
            else:
                processed_data[i, j] = 0  # 非数值替换为 0
    return processed_data

# 处理数据
processed_data = replace_non_numeric_with_zero(all_data)

# 保存处理后的数据（可选）
np.save('BJUT_all_processed.npy', processed_data)
```
